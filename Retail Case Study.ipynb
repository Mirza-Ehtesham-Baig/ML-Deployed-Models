Dealer Name vs Order Recd\Lost

```python
orders_conf=pd.crosstab(retail_data["Dealer Name"],retail_data[" Order Recd\Lost "])
orders_conf
```

```python
Order_Received=pd.DataFrame(orders_conf["Order Received"])
Order_Received
```

```python
Order_Received.sum()/len(retail_data)*100
# Success Pct
```

```python
100-Order_Received.sum()/len(retail_data)*100
# Lost Pct
```

```python
orders_conf.drop("Order Received",axis=1,inplace=True)
orders_conf
```

```python
losss=pd.DataFrame(orders_conf.sum(axis=1),columns=["lost_count"])
losss
```

```python
count=pd.merge(left=Order_Received,right=losss,how="inner",on="Dealer Name")
count["total"]=count.sum(axis=1)
count
```

```python
count["Order_Received_percentage"]=count["Order Received"]/count["total"]*100
count
```

```python
count["Order_lost_percentage"]=count["lost_count"]/count["total"]*100
count
```

```python
# ITP is the most unsuccessful dealer in terms of orders
# Sunbird is the most successful dealer in terms of orders
```

```python
plt.figure(figsize=(10,8))

sns.countplot(data=retail_data,x="Dealer Name",hue=" Order Recd\Lost ")
```
